## Limpeza do Dataset

### Remoção de duplicatas, tratamento de dados ausentes (drop de linhas, drop de colunas).

```python
df.shape
```

```python
df.count()
```

```python
df.drop_duplicates()
```

```python
df.count()
```

Não havia duplicidade de informações, as únicas lacunas que ainda estão presentes são de informações faltantes (NaN).

```python
# Dropando colunas que são irrelevantes pra o treinamento de modelos, e possuem muita ausência de dados.
df = df.drop(['isbn','isbn13', 'link'],axis=1)
df.head(5)
```

```python
# Imputando textos genéricos para preservar colunas bookformat e genre.
df.update(df['bookformat'].fillna('Unknown'))
df.update(df['genre'].fillna('Unknown'))
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
# Checando linha com título de livro NaN.
nantitle = df[df['title'].isna()]
print(nantitle)
```

Explorando o DataFrame para verificar a importância da coluna 'img'.

```python
# Dropando a coluna de img (pois só fornece links sem utilidade, direcionando para uma aba "Bad Request") e 1 linha que tem título em branco.
df = df.dropna(subset=['title'])
df = df.drop(columns=['img'])
```

```python
df.update(df['desc'].fillna('Description not available.'))
```

```python
df.isnull().sum()
```

### Convertendo o Dataset Tratado para .csv

```python
# O comando gera o dataset limpo na área de trabalho do Collab. E é o mesmo que vou subir para a pasta do GitHub.
df.to_csv('GoodReads_cleaned.csv', index=False)
```
